```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('data of gurugram real estate.csv')


#coloumn names cleaning
data.columns = data.columns.str.strip().str.replace(' ', '_').str.lower()
print(data.columns.tolist())


#column data type cleaning
data['price'] = data['price'].str.replace('₹', '').str.replace(',', '').astype(float)
data['rate_per_sqft'] = data['rate_per_sqft'].str.replace('₹', '').str.replace(',', '').astype(int)


#categorical data cleaning
data['status'] = data['status'].str.strip().str.lower()
data['rera_approval'] = data['rera_approval'].str.strip().str.lower().map({'approved by rera': True, 'not approved by rera': False})
data['flat_type'] = data['flat_type'].str.strip().str.lower()

data.drop_duplicates()

# which is the costliest flats in dataset
costliest_flats = data.sort_values(by='price', ascending=False).head(1)
print(f"Costliest Flats: {costliest_flats['property_type'].values[0]} with price ₹{costliest_flats['price'].values[0]/100000:,.2f}")


 # which locality has highest average price
highest_average_price_by_locality = data.groupby('locality')['price'].mean().sort_values(ascending=False)
print("Average Price by Locality:", highest_average_price_by_locality.head(1))
print(f"Locality with Highest Average Price: {highest_average_price_by_locality.index[0]} with average price ₹{highest_average_price_by_locality.values[0]/100000:,.2f}")

sns.barplot(x=highest_average_price_by_locality.head(10).index, y=highest_average_price_by_locality.head(10).values)
plt.title('Average Price by Locality (Top 10)')
plt.xlabel('Locality')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.savefig('average_price_by_locality.png')
plt.show()


# which locality has highest average price per sqft'

highest_average_price_per_sqft_by_locality = data.groupby('locality')['rate_per_sqft'].mean().idxmax()
print(f"Locality with Highest Average Price per Sqft: {highest_average_price_per_sqft_by_locality} with average price per sqft ₹{data.groupby('locality')['rate_per_sqft'].mean().max():,.2f}")


# Do ready to move in flats cost more than under construction flats compare percentage difference between ready to move in and under construction flats

ready_to_move_avg_price = data[data['status'] == 'ready to move']['price'].mean()
under_construction_avg_price = data[data['status'] == 'under construction']['price'].mean()
percentage_difference = ((ready_to_move_avg_price - under_construction_avg_price) / under_construction_avg_price) * 100
print(f"Ready to move in flats cost {percentage_difference:.2f}% more than under construction flats.")


#Do rera approved flats cost more than non rera approved flats compare percentage difference between rera approved and non rera approved flats
rera_approved_avg_price = data[data['rera_approval'] == True]['price'].mean()
non_rera_approved_avg_price = data[data['rera_approval'] == False]['price'].mean()
if rera_approved_avg_price > non_rera_approved_avg_price:
    print("RERA approved flats cost more than non RERA approved flats.")
else:    print("Non RERA approved flats cost more than RERA approved flats.")


#How does area (sqft) impact property price?
sns.scatterplot(x='rate_per_sqft', y='price', data=data)
plt.title('Area (sqft) vs Price')
plt.xlabel('Area (sqft)')
plt.ylabel('Price')
plt.show()


#Which BHK configuration is the most expensive on average?
average_price_by_bhk = data.groupby('bhk_count')['rate_per_sqft'].mean().sort_values(ascending=False)
print(f"Most Expensive BHK Configuration: {average_price_by_bhk.index[0]} BHK with average price per sqft ₹{average_price_by_bhk.values[0]:,.2f}")


#Which property type (Apartment, Floor, Plot) is the costliest?
average_price_by_property_type = data.groupby('property_type')['rate_per_sqft'].mean().sort_values(ascending=False)
print(f"Costliest Property Type: {average_price_by_property_type.index[0]} with average price per sqft ₹{average_price_by_property_type.values[0]:,.2f}")


#Do certain builders or companies consistently price higher?
builder_avg_price = data.groupby('company_name')['rate_per_sqft'].mean().sort_values(ascending=False)
print(f"Builder with Highest Average Price: {builder_avg_price.index[0]} with average price per sqft ₹{builder_avg_price.values[0]:,.2f}")


#Are larger homes always more expensive per square foot?
sns.scatterplot(x='area', y='rate_per_sqft', data=data)
plt.title('Area (sqft) vs Rate per Sqft')
plt.xlabel('Area (sqft)')
plt.ylabel('Rate per Sqft')
plt.savefig('area_vs_rate_per_sqft.png')
plt.show()


```
